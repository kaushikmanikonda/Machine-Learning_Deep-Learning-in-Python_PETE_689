```python
import pandas as pd
from numpy.random import randn
import numpy as np
```

### Create a 21 ´ 4 random number matrix with column titles defines as ‘Drill, ‘No_Drill’,
### ‘Frac’, ‘No_Frac’ and name it as ‘Decision’ ( 2points).

```python
seed=100
np.random.seed(seed)
Decision=pd.DataFrame(randn(21,4),
columns='Drill No_Drill Frac No_Frac'.split())
Decision
```

### 2. Create ‘Decision_1’ matrix that includes every 3rd row of the ‘Decision’ matrix starting
### from the second row using (4 point):
### a. The ‘iloc’ function (hint: iloc[m::n] selects every nth row starting from the mth row).
### b. The ‘drop’ function.

```python
Decision_1 = Decision.iloc[2::3]
Decision_1
```

```python
Decision_1 = Decision.drop(labels=[0,1,3,4,6,7,9,10,12,13,15,16,18,19])
Decision_1
```

### 3. Create ‘Decision_2’ matrix that selects all the rows of the ‘Decision’ matrix that (4 points):
### a. The ‘Drill’ column value is positive.
### b. The ‘Drill’ and ‘Frac’ column values are positive.

```python
Decision_2 = Decision[Decision.Drill > 0]
Decision_2
```

```python
Decision_2 = Decision[(Decision.Drill > 0) & (Decision.Frac > 0)]
Decision_2
```
